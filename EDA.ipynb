In statistics, exploratory data analysis is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods.

In this project, we will be comparing various different features and their correlations with each other to find out some interesting properties of the dataset and further enhance our Data Analysis.

## Pairplot 

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv("train.csv")

X = df.drop(columns = ['Item_Outlet_Sales'])
Y = df['Item_Outlet_Sales']

df = pd.DataFrame(X, columns=df.columns)
sns.pairplot(df)
```

## Correlation Matrix

```python
corr_matrix = df.corr()
corr_matrix.style.background_gradient(cmap='coolwarm')
```

In the correlation matrix above, we can see that Item Mrp has an unusually high correlation with our target variable - Item Outlet Sales. We also see that none of the predictor variables are too highly correlated with each other and thus, this will prove beneficial to our regression models. 

### Item MRP vs Outlet Sales

We see that item MRP is positively correlated to our target variable (Item Outlet Sales). Also, 
we can see in the mrp vs target graph, items sold in supermarket 3 are cheaper than the 
ones sold in supermarket 1. Also, the items sold in the outlet location tier 3 are more 
expensive than tier 1 and 2, which are almost of the same cost. 

### ITEM VISIBILITY vs OUTLET SALE

The visibility of an item is negatively correlated to its sales price. This makes sense since the 
more an item is available, the lesser will be it’s worth. Again, we can see that items sold in 
grocery stores are the cheapest compared to the supermarkets.
